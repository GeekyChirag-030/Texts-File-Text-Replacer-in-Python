```python
from tkinter import*
from tkinter.filedialog import*
from tkinter.messagebox import*

def Open(): #Function For Opening The File 
    global Path_File 
    
    Path_File=askopenfilename(initialdir='/',title='Select the File That You Want To Open',filetypes=(("Text Files",'*.txt'),('Python Files','*.py'),("Jpg",'*.jpg'),('Png','*.png'),("All Files",'*.*')))
    
    But2=Button(Window,text="Your File is Here                                     ------> Click Me",highlightcolor="brown",activebackground="blue",activeforeground="yellow",command=Open_File,highlightthickness=10,bd=10,relief='ridge',font=("Latha",28,"bold"),bg='papayawhip',fg='black').pack(pady=30)
    
    
def Open_File(): #This Function is Call AFter When The File is OPen and In This Function Actully Data is Read from File and Placed in Text Box and Then Take Inpout From User for Replacing Text 
    global Win
    Win=Tk()
    
    global Enter
    global Data
    global T
   
    T=Text(Win,width=100,height=17,font=("Times New Roman",17,"bold"),bg="khaki")
    
    with open(Path_File) as File:
        
        Data=File.read()
    
    T.insert(INSERT,Data)
    T.pack(pady=10)
    
    Label(Win,text="What Text You Want To Replace From File and With What put In Brackets Separated By Comma?",bg="peru",fg="white",highlightthickness=3,bd=3,font=('Arial',22,"bold"),highlightcolor="black").pack(pady=10)
    
    Enter=Entry(Win,font=('Times New Roman',25,'bold'),selectbackground="blue",selectforeground="yellow",highlightcolor="black",highlightthickness=10,bd=10)
    
    
    B2=Button(Win,text="Submit",activebackground="blue",font=("Arial",25,"bold"),activeforeground="yellow",bg="brown",fg="white",command=Submit,highlightcolor="black",highlightthickness=7,bd=7,relief="sunken")
    Enter.bind("<Return>",lambda event:Submit()) #It Allows To Also Call The Submit Function When Enter Key is Pressed
    
    Enter.config(width=50)
    Enter.pack(pady=10)
     
    B2.pack(pady=20)
    Win.mainloop()
    
def Submit(): #Actully It Recieves the Data Which Entered By USer in Entry Widget and Here In This Function Replacing of Text is Done and Then Also Insert the New TExt or Data In Same Text Widget
    Info=Enter.get()
    Enter.delete(0,END)
    
    T.delete(1.0,END)
    Info=Info.split(',')
    
    User_Data=[]
    
    for J in Info: #Code To Actully Remove the Braces From String 
        Str=""
        for I in J:
            if I==')' or I=='(':
                pass
            else:
                Str=Str+I
        User_Data.append(Str)
     
    New_Data=Data.replace(User_Data[0],User_Data[1])
    T.insert(INSERT,New_Data)
    
    Ans=askyesno(title="Asking Message ?",message="Are You Want To Save These Changes Also in File ?") #Aks From User That Now These Chnages Also Save In The Same File 
    if Ans==True:
        
        with open(Path_File,"w") as NewFile:
            NewFile.write(New_Data)
            
        showinfo(title="Saved",message="Your File has been Updated with New Data ! üëç")
        
        Win.destroy()
        
    else:
        
        Win.destroy()
    
    
def Quit(): #This Function is Call When User Pressed On Quit Button 
    Ans=askyesno(title="Quit Request",message="Are You Sure You Want To Quit the Application  ?")
    
    if Ans==True:
        showinfo(title="Successfull Message!",message="Your Quit Request has been Processed üëç")
        
        Window.destroy()
    else: 
        None
```

```python
#Code for the Main Window 

Window=Tk()


Window.geometry("600x600+500+200")

Window.title(" -*-*-*-*-*-* File Text Replacer -*-*-*-*-*-*  ")
Window['bg']='#FFFFEB'
Window.iconbitmap("Icon.ico")


Fr=Frame(Window,bg="teal",highlightthickness=7,bd=7,highlightcolor='grey',cursor='hand1',relief="ridge")

La=Label(Fr,text=" * File Text Replacer * ",font=("Times New Roman",40,"bold","italic"),highlightcolor="red",highlightthickness=7,bd=7,cursor='hand2',bg='bisque',fg="black",relief="groove")

But1=Button(Window,text="Open File ",highlightthickness=15,bd=15,cursor="hand2",font=("Arial",35,"bold"),bg="brown",fg="white",highlightcolor="grey",relief="groove",activebackground="blue",activeforeground="yellow",command=Open) #Main Function Open From Where All Other Will Be Called 

But3=Button(text="Quit",fg="black",bg="red",highlightthickness=10,highlightcolor="yellow",bd=10,relief="groove",font=("Times New Roman",22,"bold"),command=Quit)

Fr.pack(pady=10)
La.pack(pady=15,padx=20)

But1.pack(pady=40)
But3.pack(pady=100,side="left",padx=20)


Window.mainloop()
```
